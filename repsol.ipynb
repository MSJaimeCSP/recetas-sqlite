### 12. Operaciones entre columnas

Crea una nueva columna llamada 'Cierre_ajustado' que **ajuste el precio de cierre por la inflación acumulada del 4.5%** (es decir, multiplica el precio de cierre por 1.045).

### 13. Cálculo de la variación diaria
Crea una nueva columna llamada 'Variacion_diaria' que calcule la variación diaria del precio de cierre respecto al día anterior. La fórmula es: Variacion_diaria = Cierre - Cierre_del_dia_anterior. Utiliza el método `shift()` para acceder al valor del día anterior.

```python
# df['Valor'].shift(1)

df['VariacionDiaria'] = (df['Close'] - df['Close'].shift(1)).round(2)  
df
```

### 14. Cálculo del porcentaje de variación diaria

Crea una nueva columna llamada `Amplitud` que refleje la volatilidad diaria, calculada como la diferencia entre el precio `Maximo` y el precio `Minimo`.

```python
df['Amplitud'] = (df['High']- df['Low']).round(2)
display(df)
```

### 19. Re-muestreo temporal (resample)

Agrupa los datos por **meses (frecuencia mensual ‘ME’)** y calcula la media del precio de **Cierre** de cada mes.

Ayuda: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html

Requiere que el DateFrame sea de tipo `DatetimeIndex`, lo cual ya hemos hecho al establecer la columna 'Date' como índice.

La **regla de frecuencia 'ME' indica que queremos agrupar por meses**, y el método `mean()` calcula la media de cada grupo mensual. El resultado es una serie temporal con la media del precio de cierre para cada mes.

Otras reglas de frecuencia comunes son:
- 'D' para días
- 'B' para días hábiles (business days)
- 'W' para semanas
- 'ME' para fin de mes (Month End)
- 'MS' para inicio de mes (Month Start)
- 'QE' para trimestres (Quarter End)
- 'YE' para años (Year End)
- 'YS' para inicio de año (Year Start)

**Cuando se reduce la frecuencia es obligatorio especificar cómo se agregan los datos dentro de cada grupo**.
**Cuando se aumenta la frecuencia, pandas rellena los nuevos índices con NaN por defecto, pero también se pueden rellenar con otros métodos como forward fill (`ffill`), backward fill (`bfill`) o `.interpolate()`**.

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df_mes = df.resample('QE', on='Date')['Close'].mean()

print(df_mes)
```
